```python
# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore') # os warning will be ignored
```

```python
# Distributions: 

vis1 = sns.displot(stats["InternetUsers"]) 

#it will show the distribution of internet users

plt.show(vis1) #it willshow the graph
```

```python
vis2 = sns.displot(stats["InternetUsers"], bins=10) 
plt.show()
```

```python
#BOX PLOTS:
vis3 = sns.boxplot(data = stats, x="IncomeGroup", y='BirthRate')
plt.show()
```

```python
vis4 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False) #lm - linear model
plt.show()
```

```python
vis6 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') 
#hue - parameter for color
plt.show()
```

```python
vis7 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = True,hue = 'IncomeGroup', height = 5 )
plt.show()
```
