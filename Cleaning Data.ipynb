```python
raw['Address'].str.split(r'[,]', expand=True)
```

```python
raw[['Blocks', 'City', 'State', 'Zip']] = raw['Address'].str.split(r'[,]', expand=True)
raw
```

```python
raw['Date'].str.split(r'[,]', expand=True)
```

```python
raw[['Day', 'Month', 'Year']] = raw['Date'].str.split(r'[ ]', expand=True)
raw
```

```python
raw[['Hours', 'Minutes', 'Seconds']] = raw['Time'].str.split(r'[ ]', expand=True)
raw
```

```python
clean = raw[['Blocks', 'City', 'State', 'Zip', 'Day', 'Month', 'Year', 'Hours', 'Minutes', 'Seconds']]
clean
```

```python
clean.to_csv('clean.csv')
```
